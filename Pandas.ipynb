# The isin() function

```python
# Q. Find the most destructive death over barsman in IPL

# Condition1 - Strike Rate = (no. of runs / no.of balls) * 100
# Condition2 - Min. 200 balls played in over no. 16-20
```

```python
# first i will apply mask on my df so that it only includes balls of over no. 16-20
delivery = pd.read_csv('deliveries.csv')
mask = delivery['over'] >= 16
death_over_delivery = delivery[mask]
death_over_delivery
```

```python
#now we will groupby batsman and get deliveries played by each batsman and then get boolean series for those who played >200 bals
all_batsmen = death_over_delivery.groupby('batsman')['batsman_runs'].count()
all_batsmen
```

```python
x= all_batsmen > 200 
x
```

```python
# now we know the size of all_batsmen and x is the same so we will filter all_batsmen using x as a mask
# we will get  a series of total balls played by each in death over such that min. 200 balls played
all_batsmen[x]
```

```python
all_batsmen[x].shape
```

```python
# now we need the name of these batsmen so we will extract them from our series (to_ list() is used to convert index object to list) 
batsmen_list = all_batsmen[x].index.to_list()
batsmen_list
```

```python
# now we will calculate strike rate of these 43 batsmen
# 1.we need runs scored by these 43 batsmen
# 2.we need balls played by these 43 batsmen
```

```python
# first from our original delivery df we need only these batsmen to be filtered here we will use isin()
delivery
```

```python
# isin() requires a list an argument ..... it gives a boolean series which we will use as a mask
mask1 = death_over_delivery['batsman'].isin(batsmen_list)
mask1
```

```python
final = death_over_delivery[mask1]
final
```

```python
# we wil calculate balls played by each batsman
balls_count = final.groupby('batsman')['batsman_runs'].count()
balls_count
```

```python
# we will calculate total runs scored by each of the batsman
total_runs = final.groupby('batsman')['batsman_runs'].sum()
total_runs
```

```python
#now for strike rate i will divide the series total_runs by balls_counts and multiply by 100
strike_rate = total_runs/balls_count * 100
strike_rate.sort_values(ascending=False)
```

# The  merge() function

```python
matches = pd.read_csv('matches.csv')
```

```python
deliveries = pd.read_csv('deliveries.csv')
deliveries
```

```python
# Q. WE want to find the most scoring batsman for each season of ipl....we will merge these two df's
# since one has season(year) and the other has batsman score info ,,,, we use id/match_id to perform merge
new = deliveries.merge(matches,left_on='match_id',right_on='id')
new
```

```python
print(deliveries.shape)
print(matches.shape)
print(new.shape)
```

```python
# now we will use groupby on new on two consecutive columns i.e., groupby season and then by batsman 
# then get the runs scored sum based on season by each batsman
# We get a MULTI-INDEX SERIES......
new.groupby(['season','batsman'])['batsman_runs'].sum() 
```

```python
# To change Multi Index Series into a data frame we use reset_index() method
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index()
```

```python
# Now as values are sorted we will use drop_duplicates() method to only get the first entry for each season
# That will be the highest scorer for each season
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first')
```

```python
#Now i can sort the above df
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first').sort_values('season')
```

# Pivot Tables

```python
deliveries = pd.read_csv('deliveries.csv')
deliveries.head(2)
```

```python
# Q. We want to create a pivot table that has teams as columns and rows as over numbers and the data as number of sixes
# so it will show which team hit how many sixes in different overs  in total we will have to refine out data first
```
